```r
data <- read.csv('C:/총_지역적인구분석2.csv')
head(data)
```

```r
inTrain <- createDataPartition(y = data$소멸지역여부, p=0.7, list=F)
train <- data[inTrain, ]
test <- data[-inTrain, ]
head(train)
head(test)
```

```r
train.data <- scale(train[, 3:6])
summary(train.data)

head(train.data)
```

```r
data.km <- kmeans(train.data, centers = 3, iter.max = 10000)
data.km$centers
```

```r
train$cluster <- as.factor(data.km$cluster)
qplot(부동산, 생활서비스, color=cluster, data=train)
```

```r
table(train$소멸지역여부, train$cluster)
```

```r
nc <- NbClust(train.data, min.nc = 2, max.nc = 15, method="kmeans")
```

```r
par(mfrow=c(1, 1))
barplot(table(nc$Best.n[1, ]))
```

```r
wssplot <- function(data, nc = 15, seed = 1234) {
    wss <- nrow(data -1) * sum(apply(data, 2, var))
    for(i in 2:nc){
        set.seed(seed)
        wss[i] <- sum(kmeans(data, center=i)$withinss)}
    plot(1:nc, wss, type="b")
    }
wssplot(train.data)
```
