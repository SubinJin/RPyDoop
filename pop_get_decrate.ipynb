# 10년 간 월, 읍면동별 인구소멸비율 구하기

### 　　　　1. for문 돌리기 전 첫 번째 파일 정리
> - 파일 불러들이기 및 필요한 열 추출(1월 여성 20~30대, 노인층 65~100세이상 :get_idx)

```python
import pandas as pd
p_0801 = pd.read_csv('C:/Users/DJA/Desktop/data/새 폴더/pop_200801_200803_u8.csv', encoding='utf-8', engine='python')
get_idx = [0,16,17,18,19,20,21,22,23,53,54,55,56]
get_idx2 = [0,85,86,87,88,89,90,91,92,122,123,124,125]
get_idx3 = [0,154,155,156,157,158,159,160,161,191,192,193,194]
# get_idx = [0,16,17,18,19,20,21,22,23,53,54,55,56,85,86,87,88,89,90,91,92,122,123,124,125,154,155,156,157,158,159,160,161,191,192,193,194]
# get_idx.append(list(range(16,24)))
# for i in range(len(p_0801.columns)):
#     print(p_0801.columns[i])
pr_0801 = p_0801[p_0801.columns[get_idx]]
# p_0801[p_0801.columns[get_idx]]
# p_0801[p_0801.columns[get_idx2]].head()
# p_0801[p_0801.columns[get_idx3]].head()
```

> - 노인 합계열 추가

```python
pr_0801['노인'] = pr_0801[pr_0801.columns[1:9]].sum(axis=1)
pr_0801.head()
```

> - 여성 합계열추가

```python
pr_0801['2030여성'] = pr_0801[pr_0801.columns[9:13]].sum(axis=1)
pr_0801.head()
```

> - 소멸비율 계산(2030 여성 / (노인층/2))

```python
pr_0801['0801소멸비율'] = pr_0801['2030여성'] / (pr_0801['노인'] / 2)
pr_0801.head()
```

> - 필요한 열만 선택(행정구역 및 소멸비율)

```python
pr_0801 = pr_0801[pr_0801.columns[[0, -1]]]
pr_0801.head()
```

> - 똑같은 작업을 2월에 대해서 진행

```python
pr_0801a = p_0801[p_0801.columns[get_idx2]]
pr_0801a['노인'] = pr_0801a[pr_0801a.columns[1:9]].sum(axis=1)
pr_0801a['2030여성'] = pr_0801a[pr_0801a.columns[9:13]].sum(axis=1)
pr_0801a['0802소멸비율'] = pr_0801a['2030여성'] / (pr_0801a['노인'] / 2)
pr_0801a = pr_0801a[pr_0801a.columns[[0, -1]]]
pr_0801a.head()
```

> - pandas merge로 행정구역에 맞게 붙임

```python
pr_0801 = pd.merge(pr_0801,pr_0801a, how='outer')
pr_0801.head()
```

### 　　　　2. 위에서 한 내용을 그대로 for문으로 실행(10년간 월, 읍면동 별 소멸비율)

```python
y_f = '{0:02d}'
m_f = '{0:02d}'

for i in range(8,18):
    for j in range(1,11,3):
        p = pd.read_csv('C:/Users/DJA/Desktop/data/새 폴더/pop_20'+
                        y_f.format(i)+m_f.format(j)+'_20'+y_f.format(i)+m_f.format(j+2)+'_u8.csv', encoding='utf-8', engine='python')
        p0 = p[p.columns[get_idx]]
        p1 = p[p.columns[get_idx2]]
        p2 = p[p.columns[get_idx3]]
        p0['노인'] = p0[p0.columns[1:9]].sum(axis=1)
        p1['노인'] = p1[p1.columns[1:9]].sum(axis=1)
        p2['노인'] = p2[p2.columns[1:9]].sum(axis=1)
        p0['2030여성'] = p0[p0.columns[9:13]].sum(axis=1)
        p1['2030여성'] = p1[p1.columns[9:13]].sum(axis=1)
        p2['2030여성'] = p2[p2.columns[9:13]].sum(axis=1)
        p0[y_f.format(i)+m_f.format(j)+'소멸비율'] = p0['2030여성'] / (p0['노인'] / 2)
        p1[y_f.format(i)+m_f.format(j+1)+'소멸비율'] = p1['2030여성'] / (p1['노인'] / 2)
        p2[y_f.format(i)+m_f.format(j+2)+'소멸비율'] = p2['2030여성'] / (p2['노인'] / 2)
        p0 = p0[p0.columns[[0, -1]]]
        p1 = p1[p1.columns[[0, -1]]]
        p2 = p2[p2.columns[[0, -1]]]
        p0 = pd.merge(p0, p1, how='outer')
        p0 = pd.merge(p0, p2, how='outer')
        pr_0801 = pd.merge(pr_0801, p0, how='outer')
pr_0801.head()
```

### 　　　　3. 지역명과 지역코드 따로 필요하기 때문에 분리 작업
> - '('기준으로 앞의 값을 지역열로 추가

```python
pr_0801['지역'] = [i.split('(')[0] for i in pr_0801[pr_0801.columns[0]]]
pr_0801.head()
```

> - '('를 기준으로 두번째 값에서 마지막 자리 제외 후 지역번호열로 추가

```python
pr_0801['지역번호'] = [i.split('(')[1][:-1] for i in pr_0801[pr_0801.columns[0]]]
pr_0801.head()
```

>- 지역번호, 지역열을 맨 앞으로 배치하고(v_idx) 여기에 소멸비율 열을 그대로 나열하는 것으로 데이터 프레임 재배치 

```python
v_idx = [-1,-2]
# tp_08_17[tp_08_17.columns[[-1, -2]]].head()
for i in range(1,len(pr_0801.columns)-2):
    v_idx.append(i)

pr_08_17 = pr_0801[pr_0801.columns[v_idx]]
pr_08_17.head()
```

### 　　　　4. 결과값 저장

```python
pr_08_17.to_csv('pr_monthly_08_17.csv',mode="w")
```
