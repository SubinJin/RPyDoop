```python
import pandas as pd
p_0801 = pd.read_csv('C:/Users/DJA/Downloads/데이터/원본자료/새 폴더/pop_200801_200803_u8.csv')
get_idx = [0,1,70,139]
# p_0801.head()
# for i in range(len(p_0801.columns)):
#     print(p_0801.columns[i])
# tp_0801 = p_0801[['행정구역','2008년01월_계_총인구수','2008년02월_계_총인구수','2008년03월_계_총인구수']]
# tp_0801 = p_0801[p_0801.columns[get_idx]]
tp_08_17 = p_0801[p_0801.columns[[0,1]]]

tp_08_17.head()
```

```python
y_f = '{0:02d}'
m_f = '{0:02d}'

for i in range(8,18):
    for j in range(1,11,3):
        p = pd.read_csv('C:/Users/DJA/Downloads/데이터/원본자료/새 폴더/pop_20'+
                        y_f.format(i)+m_f.format(j)+'_20'+y_f.format(i)+m_f.format(j+2)+'_u8.csv')
        p = p[p.columns[get_idx]]
        tp_08_17 = pd.merge(tp_08_17,p,how='outer')

tp_08_17.head()
```

```python
tp_08_17 = tp_08_17.rename(columns = lambda x: x[2:4] + x[5:7])
tp_08_17.head()
```

```python
tp_08_17['지역'] = [i.split('(')[0] for i in tp_08_17['구역']]
tp_08_17.head()
```

```python
tp_08_17['지역번호'] = [i.split('(')[1][:-1] for i in tp_08_17['구역']]
tp_08_17.head()
```

```python
v_idx = [-1,-2]
# tp_08_17[tp_08_17.columns[[-1, -2]]].head()
for i in range(1,121):
    v_idx.append(i)

tp_08_17 = tp_08_17[tp_08_17.columns[v_idx]]
tp_08_17
```
